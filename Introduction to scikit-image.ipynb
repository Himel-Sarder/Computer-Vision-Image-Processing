# Lesson 1 — Introduction to scikit-image (skimage)

## What is **scikit-image**?

**scikit-image (skimage)** is a Python library built on:

* NumPy
* SciPy
* Matplotlib

It is mainly used for:

* Image preprocessing
* Image filtering
* Segmentation
* Feature extraction
* Research-level image analysis

It is designed to be:

* Simple
* Clean
* Research-friendly
* NumPy-based

## skimage vs OpenCV

| Feature                | OpenCV          | scikit-image |
| ---------------------- | --------------- | ------------ |
| Speed                  | Very Fast (C++) | Moderate     |
| Research Friendly      | Medium          | Very High    |
| Image Type             | BGR             | RGB          |
| Deep Learning Usage    | Very common     | Less         |
| Medical Image Research | Less common     | Very common  |

## Reading an Image

### Method 1: Read from file

````python
from skimage import io

image = io.imread("image.png")
````

### What does it return?

* A NumPy array
* Shape example:
  * Grayscale → `(height, width)`
  * RGB → `(height, width, 3)`

### Check shape:

````python
print(image.shape)
print(image.dtype)
````

### Displaying an Image

````python
plt.imshow(image)
# plt.axis("off")
plt.show()
````

Unlike OpenCV:
* No `cv2.imshow`
* No need for waitKey
### Image Data Format
In OpenCV:
* Image range → `0 to 255`
* dtype → `uint8`

In skimage:

* Many functions expect → `0 to 1` (float)

### Convert to float:

````python
from skimage import img_as_float

image_float = img_as_float(image)
````

````python
print(image_float.dtype)
````

Now:
* dtype → float64
* Range → 0.0 to 1.0
This is VERY important for filters.

````python
print(image.min(), image.max())
````

````python
print(image_float.min(), image_float.max())
````

### What does **range 0 to 1** mean?

It means pixel values are stored as **decimal (float) numbers between 0.0 and 1.0** instead of integers between 0 and 255.
### Why Convert to 0–1?

Because many skimage functions expect normalized data.

Mathematically:

```
new_value = old_value / 255
```

Example:

| Original (0–255) | Converted (0–1) |
| ---------------- | --------------- |
| 0                | 0.0             |
| 128              | 0.5019          |
| 255              | 1.0             |


### Why This Matters (Very Important)

If you apply a filter on 0–255 image when it expects 0–1:

* You may get wrong contrast
* Brightness may blow out
* Results may look strange

Especially important in:

* Medical images
* Thresholding
* Segmentation
* Feature extraction

### Loading Built-in Sample Images
Skimage provides built-in images for practice.

````python
from skimage import data

camera = data.camera()      # grayscale
astronaut = data.astronaut()  # RGB

plt.imshow(camera, cmap="gray")
plt.axis("off")
plt.show()
````

````python
plt.imshow(astronaut, cmap="gray")
plt.axis("off")
plt.show()
````

### Understanding Image as NumPy Array

````python
print(type(image))   # numpy.ndarray
print(image[0,0])    # pixel value
````

For RGB:

````python
print(astronaut[0,0,0])  # Red channel
print(astronaut[0,0,1])  # Green channel
print(astronaut[0,0,2])  # Blue channel
````

### Basic Grayscale Conversion

````python
from skimage.color import rgb2gray

gray = rgb2gray(image)
plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
````
